## Data Preprocessing - Dropping car_ID

```python
df_drop_id = df.drop(['car_ID'], axis=1) # Remove ID column since its not useful information
print(df_drop_id.shape)
df_drop_id.head()
```

## Data Preprocessing - Catagorical data mapping

```python
df_categorical_map = df_drop_id.copy() # Checkpoint
```

```python
catagorical_features = df_categorical_map.select_dtypes(include='object') # Obtain all categorical columns
print(catagorical_features.shape)
catagorical_features.head()
```

```python
encoder = ce.OrdinalEncoder(cols=catagorical_features) # Using Ordinal encoder
# encoder = ce.OneHotEncoder(cols=catagorical_features) # Using One Hot econder
encoded_features = encoder.fit_transform(df_categorical_map)
encoded_features.head()
```

## Data Preprocessing - Scaling

```python
from sklearn.preprocessing import StandardScaler

# Scaling data so all features have similar importance numerically
df_scaled = encoded_features.copy()
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_scaled)
df_scaled = pd.DataFrame(data=df_scaled, columns=encoded_features.columns)
df_scaled.head()
```

## Data Preprocessing - Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_scaled.drop(['price'], axis=1) # Drop target column to obtain just features
y = df_scaled['price']
x_train, x_test, y_train, y_test = train_test_split(X, y, train_size=.80, random_state=42) # 80/20 split
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(x_test.shape)
```

## Data Modeling

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
model.score(x_train, y_train)
```

```python
model.score(x_test, y_test)
```
